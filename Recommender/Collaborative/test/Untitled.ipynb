```python
destination_id_list=[i for i in range(len(destinations))]
destinations['id']=destination_id_list
destinations.set_index('id')
```

```python
faulty_id_list=list(destinations[destinations['avg_rating']=='N']['id'])
```

```python
faulty_id_list
```

```python
for i in faulty_id_list:
    destinations.drop(i,inplace=True)
```

```python
destinations
```

```python
destinations=destinations[:1000]
destinations.reset_index(inplace=True,drop=True)
destinations['voted_by']=pd.to_numeric(destinations['voted_by'])
destinations['avg_rating']=pd.to_numeric(destinations['avg_rating'])
```

```python
# destinations.to_csv('only_rated_')
```

```python
C = destinations['voted_by'].mean()
m = destinations['avg_rating'].mean()
C,m
```

```python
def bayesian_avg(avg_votes,voted_by):
    bayesian_avg = (C*m+avg_votes*voted_by)/(C+voted_by)
    return bayesian_avg
```

```python
bayesian_avg_list=[]
for i in range(len(destinations)):
#     print(destinations.loc[i]['avg_rating'])
#     print(destinations.loc[i]['voted_by'])
    bayesian_avg_list.append(bayesian_avg(destinations.loc[i]['avg_rating'],destinations.loc[i]['voted_by']))
bayesian_avg_list
```

```python
destinations['bayesian_avg']=bayesian_avg_list
```

```python
destinations.drop('id',axis=1)
```

```python
destinations.to_csv("top_1000_destinations_of_nepal_bayesian_avg.csv",index=None)
```
